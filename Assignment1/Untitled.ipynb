```python
# %load MakeDataset.py
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
#
#                                   ES335- Machine Learning- Assignment 1
#
# This file is used to create the dataset for the mini-project. The dataset is created by reading the data from
# the Combined folder. The data is then split into training, testing, and validation sets. This split is supposed
# to be used for all the modeling purposes.
#
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

# Library imports
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import os

# Constants
time = 10
offset = 100
folders = ["LAYING","SITTING","STANDING","WALKING","WALKING_DOWNSTAIRS","WALKING_UPSTAIRS"]
classes = {"WALKING":1,"WALKING_UPSTAIRS":2,"WALKING_DOWNSTAIRS":3,"SITTING":4,"STANDING":5,"LAYING":6}

combined_dir = os.path.join("Combined")

#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
                                                # Train Dataset
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

X_train=[]
y_train=[]
dataset_dir = os.path.join(combined_dir,"Train")

for folder in folders:
    files = os.listdir(os.path.join(dataset_dir,folder))

    for file in files:
        df = pd.read_csv(os.path.join(dataset_dir,folder,file),sep=",",header=0)
        df = df[offset:offset+time*50]
        X_train.append(df.values)
        y_train.append(classes[folder])

X_train = np.array(X_train)
y_train = np.array(y_train)
print(X_train.shape)
print(y_train.shape)


#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
                                                # Test Dataset
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

X_test=[]
y_test=[]
dataset_dir = os.path.join(combined_dir,"Test")

for folder in folders:
    files = os.listdir(os.path.join(dataset_dir,folder))
    for file in files:

        df = pd.read_csv(os.path.join(dataset_dir,folder,file),sep=",",header=0)
        df = df[offset:offset+time*50]
        X_test.append(df.values)
        y_test.append(classes[folder])

X_test = np.array(X_test)
y_test = np.array(y_test)

#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
                                                # Final Dataset
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=

# USE THE BELOW GIVEN DATA FOR TRAINING and TESTING purposes

# concatenate the training and testing data
X = np.concatenate((X_train,X_test))
y = np.concatenate((y_train,y_test))

# split the data into training and testing sets. Change the seed value to obtain different random splits.
seed = 4
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=seed,stratify=y)

print("Training data shape: ",X_train.shape)
print("Testing data shape: ",X_test.shape)

#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-==-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=


```

```python
import tsfel
train_tsfel_features_list = []

# Loop through each window in X_train
for i in range(X_train.shape[0]):
    # Flatten the current window and create a DataFrame
    train_data_for_tsfel = pd.DataFrame({
        'x_comp': X_train[i, :, 0],
        'y_comp': X_train[i, :, 1],
        'z_comp': X_train[i, :, 2]
    })

    # Get TSFEL configuration for statistical features
    cfg = tsfel.get_features_by_domain('statistical')

    # Extract features for the current window
    train_features = tsfel.time_series_features_extractor(cfg, train_data_for_tsfel, fs=50, window_size=200)
    
    # Append the extracted features to the list
    train_tsfel_features_list.append(train_features)

# Combine the features from all windows into a single DataFrame
train_tsfel_features_df = pd.concat(train_tsfel_features_list, ignore_index=True)

# Standardize the features
scaler = StandardScaler()
train_standardized_tsfel_features = scaler.fit_transform(train_tsfel_features_df)

# Display the shape of standardized features
print(train_standardized_tsfel_features.shape)
```

```python
# Proceed with PCA or any other analysis as needed
from sklearn.decomposition import PCA
pca_tsfel=PCA(n_components=2)
pca_tsfel.fit(train_standardized_tsfel_features)
x_pca_tsfel=pca_tsfel.transform(train_standardized_tsfel_features)
plt.figure(figsize=(10, 8))

for activity, activity_code in classes.items():
    rows=np.where(y_train == activity_code)
    plt.scatter(x_pca_tsfel[rows, 0], x_pca_tsfel[rows, 1], label = activity)
    plt.xlabel('First principle component')
    plt.ylabel('Second principle component')
    plt.title('Principle Component Analysis')
plt.legend()
plt.show()
print(train_standardized_tsfel_features.shape)
print(y_train.shape)
```

Question 4:

Calculate the correlation matrix of the features obtained by TSFEL and provided in the dataset. Identify the features that are highly correlated with each other. Are there any redundant features? [1 marks]

```python
# Calculate and display the correlation matrix
#tsfel  features:

correlation_matrix_tsfel = train_tsfel_features_df.corr()
print(correlation_matrix_tsfel.shape)
print("Correlation Matrix for tsfel Features: ")
print(correlation_matrix_tsfel)
```

```python
# Calculate and display the correlation matrix
#dataset  features:

correlation_matrix_raw = X_train_dataset.corr()

print("Correlation Matrix for Raw Features:")
print(correlation_matrix_raw.shape)
print(correlation_matrix_raw)
```

```python
def get_highly_correlated_pairs(correlation_matrix, threshold=0.8):
    # Get the absolute value of the correlation matrix
    abs_corr_matrix = correlation_matrix.abs()

    # Find pairs of features with correlation higher than the threshold
    highly_correlated_pairs = []
    for i in range(abs_corr_matrix.shape[0]):
        for j in range(i):
            if abs_corr_matrix.iloc[i, j] > threshold:
                feature1 = abs_corr_matrix.columns[i]
                feature2 = abs_corr_matrix.columns[j]
                highly_correlated_pairs.append((feature1, feature2, abs_corr_matrix.iloc[i, j]))

    return highly_correlated_pairs
```

```python
# Identify highly correlated features in TSFEL features
highly_correlated_tsfel = get_highly_correlated_pairs(correlation_matrix_tsfel)
print("Highly Correlated Features in TSFEL Features:")
print(highly_correlated_tsfel)
```

```python
# Identify highly correlated features in Raw Dataset features
highly_correlated_raw = get_highly_correlated_pairs(correlation_matrix_raw)
print("Highly Correlated Features in Raw Dataset Features:")
print(highly_correlated_raw)
```

```python
# Heatmap for TSFEL Features
import seaborn as sns
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_tsfel, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix for TSFEL Features')
plt.show()
```

```python
# Heatmap for Raw Dataset Features
import seaborn as sns
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_raw, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix for Raw Dataset Features')
plt.show()
```
